# **`DECISION TREE CLASSIFICATION WITH POST PRUNING AND PRE PRUNING`**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
df = sns.load_dataset("iris")
```

```python
dataset.target
```

```python
# Independent and dependent features
X = df.iloc[:,:-1]
y = dataset.target
```

```python
## DO NOT DO FEATURE SCALING IN DECISION TREE
```

```python
# train_test split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion="entropy")
```

```python
classifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

```python
## Post pruning technique
classifier = DecisionTreeClassifier(criterion="entropy",max_depth=2)
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

```python
## prediction
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
score = accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

## DecisionTree Prepruning and Hyperparameter tuning for Huge data


```python
import warnings
warnings.filterwarnings("ignore")
```

```python
parameter = {
   "criterion":['gini', 'entropy', 'log_loss'],
    "splitter" : ["best","random"],
    "max_depth" :[1,2,3,4,5],
    "max_features" : ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = DecisionTreeClassifier()
clf = GridSearchCV(classifier,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
score = accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
