# **`MATPLOTLIB`**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c="#223BDC",alpha=0.9,marker="X")
plt.xlabel("this is my x-axis")
plt.ylabel("this is my y-axis")
plt.title("this is my x vs y")
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,"--r")
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sine wave graph")
plt.grid()
```

```python
x = ["a","b","c","d","e"]
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color= "#7E10D5")
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color= "#7E10D5")
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
x = [3,4,5,6,7,8]
y=[4,5,6,7,8,9]
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y)
```

```python
data = [1,2,3,4,5,6,7,6,4,3,3,4,3,3,3,6,6,7]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data,color="#700BE0")
plt.xlabel("representing unique val")
plt.ylabel("representing frequecny")
plt.title("Hist plot example")
plt.grid()
plt.savefig("test.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax =fig.add_subplot(projection = "3d")
ax.scatter(x,y,z,c="#9327E8",alpha = 0.8)
plt.xlabel("my x-axis")
plt.ylabel("my y-axis")
plt.show()
```
