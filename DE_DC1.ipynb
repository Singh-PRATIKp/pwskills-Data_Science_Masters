# Drop Missing Values

- DataFrame.drop(labels=None, axis=0, columns=None)
- DataFrame.dropna()

Parameters:

- labels: A list of labels (row indices or column names) to drop.
- axis: Specifies whether to drop rows (0) or columns (1).
- columns: A list of column names to drop.


| Method                                      | Purpose                                      |
|---------------------------------------------|----------------------------------------------|
| df.dropna()                                 | Remove rows with any NaN                     |
| df.dropna(how='all')                        | Remove rows where all columns are NaN        |
| df.dropna(axis=1)                           | Remove columns with any NaN                  |
| df.dropna(axis=1, how='all')                | Remove columns where all values are NaN      |
| df.dropna(subset=['column_name'])           | Remove rows where a specific column is NaN   |
| df.dropna().reset_index(drop=True)          | Remove NaN rows and reset index              |



```python
import pandas as pd
import numpy as np

# Creating a sample DataFrame with missing values
data = {
    'Order ID': ['O-1001', 'O-1002', 'O-1003', 'O-1004', np.nan],
    'Customer ID': ['C-001', 'C-002', np.nan, 'C-004', np.nan],
    'Sales': [250.0, np.nan, 300.0, 400.0, np.nan],
    'Profit': [50.0, 20.0, np.nan, 80.0, np.nan],
    'Null Column Test' : [np.nan,np.nan,np.nan,np.nan,np.nan]
}

df = pd.DataFrame(data)
df
```

```python
df.dropna() #Drop All Rows with Any Missing
#Removes all rows where any column has NaN.
```

```python
#Drop Rows Only If All Columns Are NaN
#Removes a row only if every column in that row is NaN.
df.dropna(how='all')
```

```python
#Drop Columns with Any Missing Values
#Removes all columns that have at least one NaN.
df.dropna(axis=0) #drop rows
df.dropna(axis=1) #drop colums
#it's drop all the rows bcs last row of every column was nan
```

```python
#Drop Columns Only If All Values Are NaN
# Removes columns only if every row is NaN in that column.
df.dropna(axis=1, how='all')
```

```python
df
```

```python
#Drop Rows Where a Specific Column Has NaN
 #Removes rows where the Sales column is NaN, but keeps other rows.
df.dropna(subset = ['Sales'])
```

```python
#Reset Index After Dropping Rows
df.dropna().reset_index(drop=True)
```

## Filling Data in NAN

| Method | Purpose |
|--------|---------|
| `df.fillna(0)` | Fill NaN with 0 |
| `df.fillna(df.mean())` | Fill NaN with column mean |
| `df.fillna(method='ffill')` | Fill NaN with the previous rowâ€™s value |
| `df.fillna(method='bfill')` | Fill NaN with the next rowâ€™s value |
| `df.interpolate()` | Fill NaN by smart guessing (linear interpolation) |
| `df.replace(-999, np.nan)` | Convert placeholder values (-999 â†’ NaN) |
| `df.replace({old_value: new_value})` | Replace specific values |
| `df.loc[df['col'] < 0, 'col'] = 0` | Replace based on a condition |


```python
#Step 1: Creating Sample Data with Missing & Incorrect Values
import pandas as pd
import numpy as np

# Creating a sample dataset
data = {
    'Order ID': ['O-1001', 'O-1002', 'O-1003', 'O-1004', 'O-1005'],
    'Customer ID': ['C-001', 'C-002', 'C-003', 'C-004', 'C-005'],
    'Sales': [250.0, np.nan, 300.0, np.nan, 400.0],  # Missing Sales
    'Profit': [50.0, -999, np.nan, 80.0, 0],  # Incorrect values (-999, 0)
    'Ship Mode': ['Same Day', 'Second Class', np.nan, 'Standard Class', 'Same Day']  # Missing Ship Mode
}

df = pd.DataFrame(data)

print("ðŸ”¹ Original DataFrame:")
df
```

```python
#Fill Missing Values Using .fillna()
#âœ… Fill NaN with a Fixed Value
#ðŸ“Œ Use Case: If missing values should be considered as zero.
df.fillna(0)
```

```python
#âœ… Fill NaN with Column Mean
#ðŸ“Œ Use Case: If missing values should be replaced with the average sales.
df['Sales'].fillna(df['Sales'].mean())
```

```python
df
```

```python
#Forword Fill
#âœ… Fill Using Previous Rowâ€™s Value (ffill)
#ðŸ“Œ Use Case: If sales follow a trend over time, forward-fill (ffill) takes the last known value.
df['Sales'].fillna(method='ffill')
```

```python
#Backword Fill
#âœ… Fill Using Next Rowâ€™s Value (bfill)
#ðŸ“Œ Use Case: If sales remain stable, backward-fill (bfill) takes the next known value
df['Sales'].fillna(method='bfill')
```

```python
#âœ… Fill Using Interpolation (Smart Guessing)
#ðŸ“Œ Use Case: If data is continuous, interpolation estimates missing values smoothly.
df['Sales'].interpolate()
```

```python
#ðŸ”¹ 2ï¸âƒ£ Replace Incorrect Values Using .replace()
#âœ… Replace Specific Values
print(df)
df['Profit'].replace(-999, np.nan)
```

```python
#âœ… Replace Multiple Values at Once
#ðŸ“Œ Use Case: Convert category names (e.g., "Same Day" â†’ "Express").
print(df)
df['Ship Mode'].replace({'Same Day': 'Express', 'Second Class': 'Standard'})
```

```python
#âœ… Replace Using a Condition
#ðŸ“Œ Use Case: Fix incorrect values based on a condition.
print(df)
df.loc[df['Profit'] < 0, 'Profit'] = 0  # Set negative profits to 0
```

###  Handling Date Inconsistencies with to_datetime()

 - to_datetime(): function is used to convert DataFrame column to a datetime data type.

- Syntax:

- DataFrame ['Column'] = pd.to_datetime(arg , format)
Parameters:

- arg: The DataFrame column to convert.
- format: Specifies the format of the date strings (e.g., '%Y-%m-%d' for YYYY-MM-DD).
- Note:

- Extract components: dt accessor can be used to extract the desired date components.

- dt.year: Extracts the year.
- dt.month: Extracts the month.
- dt.day: Extracts the day.

```python
import pandas as pd

# Sample DataFrame with mixed date formats
df = pd.DataFrame({
    'Order Date': ['2024-03-01', '03/02/2024', 'March 3, 2024', '04-04-2024', 'Invalid Date']
})
df
```

```python
df.dtypes
```

```python
# Convert to datetime (force errors to NaT)
# pd.to_datetime(df['Order Date'], format='%Y-%m-%d')
pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df['Order Date'].dt.year
df['Order Date'].dt.month
df['Order Date'].dt.day
```
