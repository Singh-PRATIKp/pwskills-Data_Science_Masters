```python
#How to get number of columns.
print(df_jcs.shape[1])
print(len(df_jcs.columns))
```

```python
#using filter
## Select columns 'Customer ID', 'Customer Name', and 'City'
df_jcs.filter(items=['Customer ID', 'Customer Name','City']) #by default axis = 1
```

```python
# df_jcs.filter(items=['Customer ID', 'Customer Name','City'],axis =0)
df_jcs.filter(items=['Customer ID', 'Customer Name','City'],axis =1)
```

```python
# Select columns that contain the substring 'Name'
df_filtered1 = df_jcs.filter(like='Name', axis=1) # 'Customer Name'
print(df_filtered1.head())
```

```python
# Select columns that start with 'Cust' and end with 'ID'
df_filtered2 = df_jcs.filter(regex='^Cust.*ID$', axis=1) # 'Customer ID'
print(df_filtered2.head())
```

### Regular Expressions (Regex)

```python
# printing all customers with names terminating with the substring: 'an'
dfc['Customer Name'].str.contains('an$')
dfc[dfc['Customer Name'].str.contains('an$')]
```

```python
# customer names starting with 'Li' and ending with 'an'
dfc['Customer Name'].str.contains('^Li.*an$')
dfc[dfc['Customer Name'].str.contains('^Li.*an$')]
```

```python
# printing records by the states with names starting with "Pen"
# dfc[dfc['State'].str.contains('^Pen')]
dfc[dfc['State'].str.slice(0,3)=='Pen']
```
