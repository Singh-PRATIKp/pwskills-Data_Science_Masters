#  CONDITIONAL STATEMENTS IN PYTHON [IF-ELIF-ELSE]

 Conditional statements is based on the condition of the expressions. Conditional statements  decides whether certain statements need to be executed or not.

## Indentation 

Indentation is the white spaces at the beginning of a code.In python indentation is very important because python realise on indentation .Python uses indentation to indicate a block of code.

## IF Keyword 

Python if Statement is used for decision-making operations. It contains a body of code which runs only when the condition given in the if statement is true.

```python
a = 33
b = 200
if b > a:
  print("b is greater than a")
```

## ELIF Keyword 

Python Elif Statement is used for decision-making operations.The elif keyword is pythons way of saying "if the previous conditions were not true, then try current condition".

```python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

## ELSE Keyword 

Python else statement is used for decision-making operations.The else keyword executes anything which isn't executed by the preceding conditions.
Else is executed only when all the previous If & Elif conditions are fail.

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

1. You can also have an else without the elif.
2. There can be any number of elif statement.
3. Else is executed only when all the previous If & Elif conditions are fail.

## Short Hand If 

If you have only one statement to execute, you can put it on the same line as the if statement.

```python
if a > b: print("a is greater than b")
```

## Short Hand If-Else 

If you have only one statement to execute, one for if, and one for else, you can put it all on the same line:

```python
#One line if else statement:

a = 2
b = 330
print("A") if a > b else print("B")
```

```python
#One line if else statement, with 3 conditions:

a = 330
b = 330
print("A") if a > b else print("=") if a == b else print("B")
```

## Nested If

You can have if statements inside if statements, this is called nested if statements.

```python
x = 41

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```

Exercise 35: Dog Years

It is commonly said that one human year is equivalent to 7 dog years. However this
simple conversion fails to recognize that dogs reach adulthood in approximately two
years. As a result, some people believe that it is better to count each of the first two
human years as 10.5 dog years, and then count each additional human year as 4 dog
Write a program that implements the conversion from human years to dog years
described in the previous paragraph. Ensure that your program works correctly for
conversions of less than two human years and for conversions of two or more human
years. Your program should display an appropriate error message if the user enters
a negative number.


```python
human_year=int(input("Enter the human years : "))
if (human_year>=0):
    if (human_year<=2):
        dog_year=(human_year * (10.5/2))
    else:
        dog_year=10.5 + ((human_year-2)*4)
    print(f'Human_year: {human_year} = Dog_year : {dog_year}')

else:
    print("Please enter correct human year.")
```

Exercise 40: Name that Triangle

A triangle can be classified based on the lengths of its sides as equilateral, isosceles
or scalene. All 3 sides of an equilateral triangle have the same length. An isosceles
triangle has two sides that are the same length, and a third side that is a different
length. If all of the sides have different lengths then the triangle is scalene.
Write a program that reads the lengths of 3 sides of a triangle from the user.
Display a message indicating the type of the triangle.


```python
len1=float(input("Enter the length of first side of a triangle : "))
len2=float(input("Enter the length of second side of a triangle : "))
len3=float(input("Enter the length of third side of a triangle : "))
if (len1==len2 and len2==len3 and len3==len1):
    print("Equilateral")
elif(len1!=len2 and len2!=len3 and len3!=len1):
    print("Scalene")
else:
    print("Isosceles")
```
