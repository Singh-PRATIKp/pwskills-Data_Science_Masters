# Keywords

- Keywords are predefined, reserved words used in Python programming that have special meanings to the compiler.
- Keywords are used to define the syntax and structure of the Python language.
- We cannot use a keyword as a variable name, function name, or any other identifier.
- All the keywords except True, False and None are in lowercase and they must be written as they are.

```python
# List of keywords.
import keyword
print(keyword.kwlist)
```

# Identifiers

- Identifiers are the name given to variables,function, classes, methods, etc.
- language = 'Python'
- Here, language is a variable (an identifier) which holds the value 'Python'.

#### Rules for Naming an Identifier
- Identifiers cannot be a keyword.
- Identifiers are case-sensitive.
- It can have a sequence of letters and digits. However, it must begin with a letter or _. The first letter of an  identifier cannot be a digit.
- It's a convention to start an identifier with a letter rather _.
- Whitespaces are not allowed.
- We cannot use special symbols like !, @, #, $, and so on.

## Break

Break statement use to come out of the loop immediately .Break statement terminate the current loop.

```python
for i  in range(11):
    print(i,end=" ")
```

```python
for i in range(11):
    if i ==6:
        break
    print(i,end=" ")
```

```python
#Searching is an example of break.
#Input:pratiksingh201811@gmail.com
#Output:pratiksingh201811
email=input("Enter your E-mail ID : ")
for c in email:
    if (c=='@'):
        break
    print(c, end='')
```

```python
lower=int(input("Enter the lower range :"))
upper=int(input("Enter the upper range :"))
for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
```

## Continue

Skip the current iteration of a loop and move to the next iteration

```python
for i in range(11):
    if i == 5:
        continue
    print(i,end=" ")
```

```python
#Input:pratiksingh201811@gmail.com
#Output:pratiksingh201811
#       gmail.com
email=input("Enter your E-mail ID : ")
for c in email:
    if (c=='@'):
        print("")
        continue
    print(c, end='')
```

## Pass

Do nothing. Ignore the condition in which it occurred and proceed to run the program as usual

```python
for i in range(11):
    if (i%3==0):
        print(i,end=" ")
    else:
        pass
```

## Escape Character

Escape characters is a mechanism in python which allows us to insert characters that are illegal in a python string.
An escape character is a backslash \ followed by the character you want to insert.

\r- Carriage Return takes the cursor to the beginning of the current line.
\n-Newline takes the cursor to the next line.

```python
print('It's a beutiful day.')
```

```python
print("It's a beutiful day.")
print('It\'s a beutiful day.')
```

```python
print('We\'re from "IIT madras".')
```

```python
#Tab=\t
print("My name is pratik. I am from Mumbai.")
print("My name is pratik.\tI am from Mumbai.")
print("My name is pratik.\t\t I am from Mumbai.")
```

```python
#newline=\n
print("My name is pratik.\nI am from Mumbai.")
```

```python
print("My name is pratik singh, I'm from mumbai.")
print('My name is pratik singh, I\'m from mumbai.') #single quote \'
print("My name is pratik singh, I\\m from mumbai.")  #backslash  \\
print("My name is pratik singh,\n I'm from mumbai.")  #newline \n
print("My name is pratik singh,\r I'm from mumbai.")  #carriage return \r
print("My name is pratik singh, \t I'm from mumbai.")  #tab \t
print("My name is pratik singh, \bI'm from mumbai.")   #backspace \b
```
