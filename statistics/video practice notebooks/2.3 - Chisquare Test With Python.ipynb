# **`CHI SQUARE TEST WITH PYTHON`**

```python
import scipy.stats as stat
import numpy as np
```

```python
## No of hours student study daily in a weekly basis
## Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday
expected_data = [8,6,7,9,6,9,7]
observed_data = [7,8,6,9,9,6,7]
```

```python
sum(expected_data),sum(observed_data)
```

```python
## Chi - square goodness of fit

chisqaure_test_statistics, p_value = stat.chisquare(observed_data,expected_data)
```

```python
print(chisqaure_test_statistics,p_value)
```

```python
## find the critical value
significance = 0.05
dof = len(observed_data) -1
critical_value = stat.chi2.ppf(1-significance,dof)
```

```python
critical_value
```

```python
if chisqaure_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```
